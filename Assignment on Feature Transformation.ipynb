```python
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MaxAbsScaler
```

# MinMaxScaler

```python
data1 = pd.read_csv('agora.csv')
```

```python
minmax=MinMaxScaler()
```

```python
data1.head(10)
```

```python
mydata=data1.drop('Profit',axis=1)
```

```python
mydata.head(10)
```

```python
mydata['Marketing Spend']=minmax.fit_transform(mydata[['Marketing Spend']])
mydata['Administration']=minmax.fit_transform(mydata[['Administration']])
mydata['Transport']=minmax.fit_transform(mydata[['Transport']])
```

```python
mydata.head()
```
